```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import talib 
from dateutil.relativedelta import relativedelta
from Tool_v1 import *
```

```python
stock_name=["AAPL","GOOG","AMZN","TSLA","BABA","PTPI","BTCY",'FOX',"NVDA","DIS","NKE","XOM","KO",
             "2319.HK","0700.HK","1211.HK",'2318.HK','1448.HK','1810.HK',"3690.HK","2018.HK"]
#stock_name="AAPL"
start="2020-12-31"
#start="2022-12-31"
#start="2023-07-30"
end="2023-09-30"
```

```python
num=[]
for i in stock_name:
    print(i)
    count=[]
    
    stock_data=pct_change(i,start,end)
    stock_data=stock_data.copy()
    
    stock_data1=MACD_short(stock_data)
    
    num_rows = stock_data1.shape[0]+1
    x=list(range(1, num_rows ,1))
    
    #print(x)
    
    y1=stock_data1["Close"]
    y2=stock_data1["DIF"]
    y3=stock_data1["MACD"]
    
    intersection_count = 0
    # 计算交点个数
    for i in range(len(x)-1 ):
        if (y2[i] - y3[i]) * (y2[i + 1] - y3[i + 1]) < 0:  # 判断是否相交
            intersection_count += 1
        
    print("6 19 交点个数：", intersection_count)
    count.append(intersection_count)
    
    stock_data2=MACD(stock_data)
    
   
    y1=stock_data2["Close"]
    y2=stock_data2["DIF"]
    y3=stock_data2["MACD"]
    
    intersection_count = 0
    # 计算交点个数
    for i in range(len(x)-1 ):
        if (y2[i] - y3[i]) * (y2[i + 1] - y3[i + 1]) < 0:  # 判断是否相交
            intersection_count += 1
        
    print("12 26 交点个数：", intersection_count)
    
    count.append(intersection_count)
  
    stock_data3=MACD_long(stock_data)
    
   
    y1=stock_data3["Close"]
    y2=stock_data3["DIF"]
    y3=stock_data3["MACD"]
    
    intersection_count = 0
    # 计算交点个数
    for i in range(len(x)-1 ):
        if (y2[i] - y3[i]) * (y2[i + 1] - y3[i + 1]) < 0:  # 判断是否相交
            intersection_count += 1
        
    print("19 39 交点个数：", intersection_count)
    count.append(intersection_count)
    num.append(count)
    print()
    
```

```python

i=0
plt.figure(figsize=(15,10))
while i<len(stock_name):
    x=["6 19","12 26","19 39"]
    plt.plot(x,num[i],label=stock_name[i])
    i+=1

plt.legend()


plt.title("Line Chart")
plt.xlabel("Time Span")
plt.ylabel("Intersection Count")
plt.savefig('DIF_DEA.png')
plt.show()
```
